## Taking PDF File location as input, reading the file and counting the pages in the file

```python
pdf_location = input()
import PyPDF2
from PyPDF2 import PdfFileReader
pdfFileObj = open(pdf_location, 'rb') 
pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 
page_number = pdfReader.numPages  # for pages count
print(page_number)
```

## Extracting the text from given pdf file using pdfminer.highlevel

```python
from pdfminer import high_level

pdf_file = pdf_location
pages = [0]
extracted_text = high_level.extract_text(pdf_file,"",pages)
```

## Splitting the extracted data and appending into a list

```python
import re
text_list = re.split('\s+', extracted_text)
print(text_list)
```

# Finding the mobile number from extracted data

```python
#from the previous output, we can conclude the mobile_number is with Mobile text in the list

for i in range(0,len(text_list)):
    if 'Mobile' in text_list[i]:
        mobile_number = text_list[i]
mobile_number = mobile_number[0:len(mobile_number)-6]  
mobile_number = mobile_number[-10: ]
print(mobile_number)

```

# Finding the Name and Invoice period from extraced data

```python
for i in range(0,len(text_list)):
    if 'Mr' in text_list[i]:  
        break
invoice_period = text_list[i][len(text_list[i]) - 24 : len(text_list[i]) - 2]  

name = text_list[i][len(text_list[i]) - 2 : ] + ". " + text_list[i + 1] + " "
for j in range(0,len(text_list[i+2])):
    if text_list[i+2][j] >= "0" and text_list[i+2][j] <= "9":
        name += text_list[i+2][0:j]
        break
print(name)   

count = 0
for i in range(0,len(invoice_period)):
    if invoice_period[i] == "-":
        count += 1
    if count == 2:
        break

invoice_period = invoice_period[0: i+5] + " to " + invoice_period[i+5: ]
print(invoice_period)

       
```

##### Dictionary to convert Month name into Month number and vice-vera

```python
def monthToNum(shortMonth):
    return {
            'Jan': 1,
            'Feb': 2,
            'Mar': 3,
            'Apr': 4,
            'May': 5,
            'Jun': 6,
            'Jul': 7,
            'Aug': 8,
            'Sep': 9, 
            'Oct': 10,
            'Nov': 11,
            'Dec': 12
    }[shortMonth]

    
def numtoMonth(shortNum):
        return {
            1 :  'Jan',
            2 : 'Feb',
            3 : 'Mar',
            4 : 'Apr',
            5 : 'May',
            6 : 'Jun',
            7 : 'Jul',
            8 : 'Aug',
            9 : 'Sep', 
           10 : 'Oct',
           11 : 'Nov',
           12 : 'Dec'
    }[shortNum]
```

```python
from datetime import datetime

invoice_date = invoice_period[-11 : ]

day_value = int(invoice_date[0:2])
month_value = invoice_date[3:-5]
year_value = int(invoice_period[-4 : ])
month_value = monthToNum(month_value)
```

```python
# Invoice issued date, because address is paired up with the invoice issued date. 

import datetime
from datetime import timedelta

today_date = datetime.datetime(year_value,month_value,day_value)
next_date = today_date + timedelta(days = 1) 
next_date = str(next_date)
year_value_n = next_date[0:4]
month_value_n = next_date[5:7]
day_value_n = next_date[8: 10]
month = numtoMonth(int(month_value_n))
invoice_recieved_date = day_value_n + "-" + month + "-" + year_value_n
print(invoice_recieved_date)
```

## Getting address from the extracted data

```python
for i in range(0,len(text_list)):
    if text_list[i] == 'Value':   # address of the user are present after Value, in the list
        break
for j in range(i,len(text_list)):
    if invoice_recieved_date in text_list[j]:  # the last line of address contains, invoice issued date in the extracted text
        break 
address = ""

for z in range(i+1,j):
    address += text_list[z] + " "

flag = 0
for i in range(0,len(text_list[j])):
    if (text_list[j][i] >= "A" and text_list[j][i] <= "Z") or (text_list[j][i] >= "a" and text_list[j][i] <= "z"):
        flag = 1
    if flag == 1 and (text_list[j][i] >= "0" and text_list[j][i] <= "9") or text_list[j][i] == ".":   
        break
address += text_list[j][0:i]
print(address)     

```

```python
#dictionary for name, address...
customer_info = {
    'Name' : name,
    'Address' : address,
    'Mobile Number' : mobile_number,
    'Invoice Period' : invoice_period
}

print(customer_info)
```

# Getting customer details from first page int JSON format

```python
import json

customer_info_json = json.dumps(customer_info, indent = 4)
print(customer_info_json)
```

## Romaing Calls

```python
def roaming_call_data(roaming_calls_list):
    for i in range(0,len(roaming_calls_list)):
        if "hh:mm:ss" in roaming_calls_list[i]:   #first roaming call for invoice period
            break
    for j in range(i+1, len(roaming_calls_list)):   #last roaming call for invoice period
        if 'Total' in roaming_calls_list[j]:
            break   
    roaming_calls = [roaming_calls_list[i][-14:]]  # first roaming call data 
    for r in range(i+1,j):
        roaming_calls.append(roaming_calls_list[r])
    roaming_calls.append(roaming_calls_list[j][0: -14])  #last roaming call data
     
    #In extracted data two roaming calls data are mixed, in adjacent list values. So, I got the each roaming call data in following way

    each_call_details = [roaming_calls[0]+roaming_calls[1]+roaming_calls[2][0:34]] 
     
    flag = 0
    for d in range(2,len(roaming_calls)-1,2):
        count = 0

        # Some mobile numbers are of 11 in length and some are of 12.
        if flag == 0:
            demo = roaming_calls[d][34 : ] + roaming_calls[d+1] + roaming_calls[d+2][0:34]
        else:
            demo = roaming_calls[d][33 : ] + roaming_calls[d+1] + roaming_calls[d+2][0:34]    
       
        if demo[-1] != '0':
            demo = demo[0:-1]  
            flag = 1
        else:
            flag = 0
    
        each_call_details.append(demo) 

    number_of_roaming_calls = len(each_call_details)

    sno = []
    date_called = []
    destination = []
    called_time = []
    number_called = []
    duration = [] 
    calltype = ['Out'] * number_of_roaming_calls #call type is always out
    total_amount = ["0.00"] * number_of_roaming_calls  # amount billed is 0.00 for every 

    # finding duraion, date, destination, number called, time called from extracted roaming calls data

    for i in range(0,number_of_roaming_calls):

        duration.append(each_call_details[i][-9:-4])  #duration
        
        # date
        for j in range(len(each_call_details[i])):
            if each_call_details[i][j] == '-':
                break
        date_called.append(each_call_details[i][j-2:j+7])

        sno.append(each_call_details[i][0:j-2])
        
        #destination
        count1 = 0
        for c in range(len(each_call_details[i])):
            if each_call_details[i][c] == '-':
                count1 += 1
            if count1 == 3:
                break  
        dest = each_call_details[i][j+7: c+1]    
        for d in range(c+1,len(each_call_details[i])):
            if each_call_details[i][d] >= '0' and each_call_details[i][d] <= '9':
                break
            else:
                dest += each_call_details[i][d]
        destination.append(dest)

        number_called.append(each_call_details[i][d+8:-12])  #Mobile_Number

        called_time.append(each_call_details[i][d:d+8]) #time called
        


        

   

    for i in range(0,len(sno),2):
        roamingcalls = {
            'S.no' : sno[i],
            'Date Called' : date_called[i],
            'Destination Called' : destination[i],
            'Time' : called_time[i],
            'Number Called' : number_called[i],
            'Type' : calltype[i],
            'Duration' : duration[i],
            'Total Amount' : total_amount[i]

        }

        roaming_call_info = json.dumps(roamingcalls, indent = 4)
        print(roaming_call_info)

    for i in range(1,len(sno),2):
        roamingcalls = {
            'S.no' : sno[i],
            'Date Called' : date_called[i],
            'Destination Called' : destination[i],
            'Time' : called_time[i],
            'Number Called' : number_called[i],
            'Type' : calltype[i],
            'Duration' : duration[i],
            'Total Amount' : total_amount[i]

        }

        roaming_call_info = json.dumps(roamingcalls, indent = 4)
        print(roaming_call_info)         



```

# Roaming calls data in JSON format

```python
#Roaming calls page
# for every page checking whether it contains Roaming call details or not.  

for page_count in range(0,page_number):
    pages = [page_count]
    extracted_text = high_level.extract_text(pdf_file,"",pages)
    if "Roaming Call Details" in repr(extracted_text):
        break
if page_count + 1 == page_number:     
    roaming_calls_list = re.split('\s+', extracted_text)
    roaming_call_data(roaming_calls_list)
else: 
    for i in range(page_count+1,page_number):
        pages = [i]
        extracted_text = high_level.extract_text(pdf_file,"",pages)
        roaming_calls_list = re.split('\s+', extracted_text)
        if "Roaming Call Details" in repr(extracted_text):
            roaming_call_data(roaming_calls_list)


```
